```python
import basedosdados as bd
```

```python
df = bd.read_sql("""
                SELECT 
                    ano, sigla_uf, causa_basica, COUNT(causa_basica) as quantidade, ocupacao, descricao_categoria
                FROM 
                    basedosdados.br_ms_sim.microdados AS sim
                    INNER JOIN  
                        basedosdados.br_bd_diretorios_brasil.cid_10 AS cid
                        ON
                            sim.causa_basica = cid.categoria
                WHERE 
                    ano = 2017
                    AND
                        ocupacao != 'None'
                GROUP BY 
                    ano, sigla_uf, ocupacao, causa_basica, descricao_categoria
                ORDER BY quantidade DESC

                LIMIT 10
            """,
            'explorando-basedosdados')
            
df

#todo: fazer a principal causa de cada estado, depois ver quais ocupações mais desviam
```
